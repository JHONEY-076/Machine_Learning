## 대구 + 빙어 시각화

```python
# 대구 + 빙어 데이터 시각화- 산포도
plt.scatter(smelt_length,smelt_weight,c="r",s=10,marker="^")
plt.scatter(bream_length,bream_weight,c="b",s=5)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
#데이터 합치기 
length=bream_length+smelt_length
weight=bream_weight+smelt_length


```

```python
len(length)
```

```python
#zip으로 데이터 묶기
zip_data=zip(length,weight)
```

```python
# 싸이킷 런에 사용할 데이터
fish_data=[]
for l,w in zip_data:
    #print("length",l)
    #print("weight",w)
    fish_data.append([l,w])
print(fish_data)
```
