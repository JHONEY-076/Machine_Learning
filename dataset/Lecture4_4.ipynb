## 특성 공학(Feature Engineering)

```python
from sklearn.preprocessing import PolynomialFeatures

poly=PolynomialFeatures(include_bias=False) ## 곱한 값, 제곱갑, 절편값의 계수 1

poly.fit([[2,3]]) ## 데이터를 집어넣어 훈련시킨다
values= poly.transform([[2,3]]) ## x0,x1을 변형 시킨다.
print(values) ## 2,3,4,6,9
names=poly.get_feature_names_out() ## 위의 변형시킨 데이터의 변경순서를 나타냄
print(names)
np.column_stack((names,values[0])) 
```
