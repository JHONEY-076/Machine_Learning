# 앙상블 분류 모델링

### RandomForest

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
wine=pd.read_csv("https://bit.ly/wine_csv_data")
data=wine[["alcohol","sugar","pH"]].to_numpy()
target=wine["class"].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input,test_input,train_target,test_target=train_test_split(data,target,test_size=0.2,random_state=42)
```

```python
### 모델링
```

```python
from sklearn.model_selection import cross_validate

from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_jobs=-1, random_state=42,n_estimators=100) ## n_estimators => tree가 100개 만들어짐을 나타냄.
```

```python
score=cross_validate(rf,train_input,train_target,return_train_score=True,n_jobs=-1)
```

```python
score["train_score"]
```

```python
score["test_score"]
```

```python
np.mean(score["train_score"])
```

```python
np.mean(score["test_score"])
```

```python
## 파라미터 최적화
```

```python
from scipy.stats import uniform, randint
params={'min_impurity_decrease':uniform(0.0001,0.001),
        'max_depth': randint(20,50),
        'min_samples_split':randint(2,25),
        'min_samples_leaf':randint(1,25),
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(rf,params,n_iter=100, n_jobs=1,random_state=42)
```

```python
rs.fit(train_input,train_target)
```

```python
best_rf=rs.best_estimator_
```

```python
best_rf.score(train_input,train_target)
```

```python
best_rf.score(test_input,test_target)
```

```python
best_rf.feature_importances_
```
