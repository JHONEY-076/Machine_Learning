## 규제(Regulation)

규제를 적용하기 전에 표준화 작업을 해줘야 한다.

```python
# StandScaler를 사용한 표준화화
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(train_poly)
```

```python
train_scaled=ss.transform(train_poly)
test_scaled=ss.transform(test_poly)
```

## 모델링

```python
from sklearn.linear_model import Ridge

ridge=Ridge() # alpha=1.0
ridge.fit(train_scaled,y_train)

print(ridge.score(train_scaled,y_train)) #학습 데이터의 성과 R2
print(ridge.score(test_scaled,y_test)) # 검증 데이터의 성과 R2

```

```python
import matplotlib.pyplot as plt

# 변수 초기화 - 성과지표 담을 리스트
train_scores=[]
test_scores=[]

# alpha 리스트 => Ridge 규제 강도 조절 파라미터
alpha_list=[0.001,0.01,0.1,1,10,100]
for alpha in alpha_list:
    print("alpha: ",alpha)
    ridge=Ridge(alpha=alpha)
    ridge.fit(train_scaled,y_train)
    train_scores.append(ridge.score(train_scaled,y_train))
    test_scores.append(ridge.score(test_scaled,y_test))
    print("학습:",train_scores,"검증:",test_scores)
```

```python
train_scores
```

```python
test_scores
```

```python
import matplotlib.pyplot as plt
plt.plot(np.log10(alpha_list),train_scores)
plt.plot(np.log10(alpha_list),test_scores)
plt.xlabel("log10(alpha)")
plt.ylabel("R^2")
plt.show()
```
